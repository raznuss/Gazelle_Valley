```python
import pandas as pd
import os

# File path
file_path = r"D:\Development\RESEARCH\Gazelle_Valley\Data\Precipiteion_Data\2024\ram_2024.csv"

# Load the CSV file
df = pd.read_csv(file_path)

# Convert Date Time column to datetime type
df['Date Time'] = pd.to_datetime(df['Date Time'])

# Convert Date Time column to the desired format
df['Date'] = df['Date Time'].dt.strftime('%m/%d/%Y')
df['Time'] = df['Date Time'].dt.strftime('%H:%M')

# Drop the original Date Time column
df.drop(columns=['Date Time'], inplace=True)

# Output file path
new_file_path = r"D:\Development\RESEARCH\Gazelle_Valley\Data\Precipiteion_Data\2024\txtfiles\ram_2024.csv"
output_file = os.path.join(os.path.dirname(new_file_path), "ram_2024.txt")

# Write to a text file
with open(output_file, "w") as f:
    # Write the header
    f.write(";;Name                 Date          Time         Value     \n")
    f.write(";;------------ ----------------- ------------- -------------\n")

    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        date = row['Date']
        time = row['Time']
        value = row['Rain (mm)']
        f.write(f"R01 {date} {time}    {value:.1f}\n")
    
    f.write(";;------------ ----------------- ------------- -------------\n")
```
