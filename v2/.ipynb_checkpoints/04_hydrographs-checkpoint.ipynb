# 04 — Hydrograph Visualization

Generates publication-quality dual-axis hydrographs:
- Left axis: discharge (black line)
- Right axis: Givat Ram rainfall intensity bars (inverted, blue)

Two plotting modes:
1. **By season index** — looks up the rain window from the matched events table
2. **By time range** — plots any custom start/end datetime

Both modes print an event summary (total rain, runoff volume, peak Q, runoff coefficient).

```python
import sys
from pathlib import Path
# sys.path.insert(0, str(Path('..').resolve()))

import pandas as pd

import config as cfg
from utils.discharge import load_discharge
from utils.plots import plot_hydrograph
```

## Load data

```python
discharge = load_discharge(cfg.DISCHARGE_RAW_CSV)

rain = pd.read_csv(cfg.RAM_PROC_CSV)
rain['date_time'] = pd.to_datetime(rain['date_time'])

# Load the matched events table (used for mode 1)
events_raw = pd.read_csv(cfg.FLOW_RAIN_CSV)

# The first two rows are MultiIndex header rows — skip them
events = events_raw.iloc[2:].copy()
events = events.rename(columns={
    'Unnamed: 0': 'season_index',
    'rain':       'rain_event_start',
    'rain.1':     'rain_event_end',
})
events['season_index']    = events['season_index'].astype(int)
events['rain_event_start'] = pd.to_datetime(events['rain_event_start'])
events['rain_event_end']   = pd.to_datetime(events['rain_event_end'])

print('Data loaded.')
print(f'Discharge: {len(discharge):,} rows')
print(f'Rain:      {len(rain):,} rows')
print(f'Events:    {len(events)} (season_index range: '
      f'{events["season_index"].min()} – {events["season_index"].max()})')
```

## Event list

```python
events[['season_index', 'rain_event_start', 'rain_event_end']].to_string(index=False)
```

## Mode 1 — Plot by season index

```python
def plot_event(season_index: int):
    row   = events[events['season_index'] == season_index].iloc[0]
    start = row['rain_event_start']
    end   = row['rain_event_end']
    fig   = plot_hydrograph(
        discharge_df=discharge,
        rain_df=rain,
        start=start,
        end=end,
        title=f'Hydrograph — Event {season_index}',
        catchment_area_km2=cfg.CATCHMENT_AREA_KM2,
    )
    fig.show()

# Change the number to any season_index you want
plot_event(318)
```

## Mode 2 — Plot by custom time range

```python
fig = plot_hydrograph(
    discharge_df=discharge,
    rain_df=rain,
    start=pd.Timestamp('2025-12-28 18:00'),
    end=pd.Timestamp('2025-12-29 16:00'),
    title='Hydrograph — 2025-12-28',
    catchment_area_km2=cfg.CATCHMENT_AREA_KM2,
)
fig.show()
```
